Declaring libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining constants

```python
M_1 = 3 * (10**30)
M_2 = 6 * (10**30)
AU_to_m = 1.5 * (10 ** 11) # Conversion of 1au to metres
x_1 = 45 * AU_to_m
x_2 = 135 * AU_to_m
P = 3.16 * (10 ** 9) # Period of 10 years in terms of seconds
G = 6.67 * (10 ** -11)
delta_x = 0.5 * AU_to_m
COM = (M_1 * x_1 + M_2 * x_2)/(M_1 + M_2) # Centre of Mass
omega_sq = ((2 * np.pi) / P) ** 2
```

Defining R1 and R2 (relative to centre of mass)

```python
R_1 = x_1 - COM
R_2 = x_2 - COM
```

Creating empty arrays for U(x) and the range x

```python
U_x = np.empty(361) 
x = np.empty(361)
```

Iterated loop to generate values of U(x) for plot

```python
i = 0
while i < 360:
    x[i] = -COM + delta_x * i
    if x[i] == R_1:
        x[i] = R_1 + 0.1
    if x[i] == R_2:
        x[i] = R_2 + 0.1
    U_x[i] = (-G * (M_1 * (1 / np.abs(R_1 - x[i])) + M_2 * (1 / np.abs(R_2 - x[i])))) - 0.5 * omega_sq * (x[i] ** 2)
    i += 1
```

Matplotlib functions to generate the graph with points plotted

```python
plt.figure(figsize=(20,12))
plt.plot(x[0:-2], U_x[0:-2])

plt.scatter(R_1, -0.95e9, color='red', label='Star 1', marker='*', s=250)
plt.scatter(R_2, -0.95e9, color='orange', label='Star 2', marker='*', s=250)
plt.scatter(-0.145e13, -0.098e9, color='green', label='L1 (Lagrangian Point)', marker='o', s=100)
plt.scatter(0.8e13, -0.25e9, color='blue', label='L2 (Lagrangian Point)', marker='o', s=100)
plt.scatter(-1.1e13, -0.365e9, color='black', label='L3 (Lagrangian Point)', marker='o', s=100)

plt.legend(fontsize = 'xx-large')
plt.grid(True)

plt.xlabel('Position (x)', fontsize = 'xx-large')
plt.ylabel('Potential U(x)/m', fontsize = 'xx-large')

plt.title('U(x)/m Vs. x', fontsize = 'xx-large')
plt.ylim(-1e9, 100)
plt.show()
```
